#### **DOING SET OPERATIONS IN PYTHON**

##### **Making a Set**

```python
numerical1 = set([1,2,3,4,5])
numerical2 = set([6,7,8,9,10])
```

```python
numerical1
```

```python
numerical2
```

##### **Adding and Removing**

```python
numerical1.add(9)
numerical1
```

```python
numerical2.add(12)
numerical2
```

```python
numerical1.remove(2)
numerical1
```

```python
numerical2.discard(10)
numerical2
```

```python
numerical1.pop()
numerical1
```

```python
numerical2.clear()
numerical2
```

##### **Printing all**

```python
dataScientist = set(['Python','Matplotlib','Seaborn','Power BI','SQL'])

for skill in dataScientist:
    print(skill)
```

```python
print(sorted(dataScientist))
```

```python
print(sorted(dataScientist, reverse=True))
```

##### **EXECUTION TIME**

```python
import timeit

print(timeit.timeit('list(set([1,2,6,4,4,6,1,8,9]))', number=10000))
```

#### **UNION**

```python
numerical1 = set([1,2,3,4,5])
numerical2 = set([3,4,6,2,10])

print(numerical1.union(numerical2))
```

#### **INTERSECTION**

```python
print(numerical2.intersection(numerical1))
```

##### **Difference**

```python
numerical1.difference(numerical2)
```

##### **Symmetric Difference**

```python
numerical1.symmetric_difference(numerical2)
```

##### **Set Comprehension**

```python
{skill for skill in dataScientist}
```

##### **Membership Test**

```python
'SQL' in dataScientist
```
