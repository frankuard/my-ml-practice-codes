#### **DOING SET OPERATIONS IN PYTHON**

##### **Making a Set**

```python
numerical1 = set([1,2,3,4,5])
numerical2 = set([6,7,8,9,10])
```

```python
numerical1
```

```python
numerical2
```

##### **Adding and Removing**

```python
numerical1.add(9)
numerical1
```

```python
numerical2.add(12)
numerical2
```

```python
numerical1.remove(2)
numerical1
```

```python
numerical2.discard(10)
numerical2
```

```python
numerical1.pop()
numerical1
```

```python
numerical2.clear()
numerical2
```

##### **Printing all**

```python
dataScientist = set(['Python','Matplotlib','Seaborn','Power BI','SQL'])

for skill in dataScientist:
    print(skill)
```

```python
print(sorted(dataScientist))
```

```python
print(sorted(dataScientist, reverse=True))
```

##### **EXECUTION TIME**

```python
import timeit

print(timeit.timeit('list(set([1,2,6,4,4,6,1,8,9]))', number=10000))
```

```python
print(timeit.timeit('remove_duplicates([1,2,6,4,4,6,1,8,9])',globals=globals(),number=10000))
```
