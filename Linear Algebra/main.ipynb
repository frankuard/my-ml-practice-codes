```python
import numpy as np

v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

print(v1)
print(v2)
```

#### **Vector Dot Product**

##### **Manually:**

```python
def dot(v1,v2):
    product = v1[0]*v2[0]+v1[1]*v2[1]+v1[2]*v2[2]
    return product
print(dot(v1,v2))
```

##### **Built in**

```python
print(np.dot(v1,v2))
```
